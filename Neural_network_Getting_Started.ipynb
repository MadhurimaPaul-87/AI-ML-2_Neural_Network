**What are the key parts of neural network?**

* Input
* Layer(s)
* Output

**Details**

* Feed input data in neural network
* Data flows from layer to layer until the output is generated.
* Once output is ready, calculate the error which is a scalar
* Adjust a given parameter (weight or bias) by subtracting the derivative of the error with respect to the parameter itself.
* Iterate above process.


```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Input,Dense
```

```python
input = Input(shape=(128,128,3))
```

```python
output = Dense(1)(input)
```

```python
model = tf.keras.Model(inputs=input,outputs=output)
```

```python
model.summary()
```

```python
model_name = 'simple_model.h5'
```

```python
model.save(model_name)
```
